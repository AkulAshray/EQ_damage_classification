```python
df = pd.read_csv('Clean_Data/Clean_Data/train_values.csv'); df.head()
y_labels = pd.read_csv('Clean_Data/Clean_Data/train_labels.csv')
```

```python
df, cat_col = turn_cat(df)
```

```python
# we will prepare our data for our models
df_train = pd.concat([df,y_labels], axis=1)
```

```python
# before we run our models we need to ensure that the labels are properly encoded
cat_col.append('Labels')
le = preprocessing.LabelEncoder()
for i in cat_col:
    df_train[i] = le.fit_transform(df_train[i])
```

```python
X = df_train.drop(['building_id', 'height_ft_post_eq', 'count_floors_post_eq','damage_grade', 'Labels'], axis=1)
y = df_train['Labels']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 2, shuffle = True, stratify = y)
```

## Reload our best model

```python
clf = joblib.load('gbm_third.pkl')
```

```python
lgb_prediction = clf.predict(X_test)
lgb_prediction = lgb_prediction.argmax(axis = 1)
```

```python
labels = ['low', 'medium', 'high']
plt.figure(figsize=(8,8))
plt.grid(b=False)
cm = confusion_matrix(y_test, lgb_prediction)
plot_confusion_matrix(cm, classes=labels, normalize=True, title='Normalized confusion matrix')
```

```python
def shap_plot(data, j, i=0):
    shap_values_Model = explainer.shap_values(data)
    p = shap.force_plot(explainer.expected_value[i], shap_values_Model[i][j,:], data.iloc[j,:])
    return(p)
```

```python
df_new = df_train.sample(100000)
```

```python
X = df_new.drop(['damage_grade', 'Labels'], axis=1)
```

```python
imp_features = ['geo1', 'geo2', 'geo3', 'age_building', 'has_superstructure_mud_mortar_stone', 'has_superstructure_mud_mortar_brick',
               "has_superstructure_rc_non_engineered", 'height_ft_pre_eq', 'has_superstructure_timber']
X = X[imp_features]
```

```python
X.shape
```

```python
y = df_new['Labels']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 2, shuffle = True, stratify = y)
```

```python
train_data_lgbm = lgbm.Dataset(X_train, label=y_train)
test_data_lgbm = lgbm.Dataset(X_test, label=y_test)

learning_rate = 0.05
num_leaves = 20
feature_fraction = 0.6

params = {"objective": "multiclass",
          "boosting_type": "gbdt",
          "learning_rate": learning_rate,
          "num_leaves": num_leaves,
           "max_bin": 256,
          "feature_fraction": feature_fraction,
          "task": 'train',
          "valid_sets": [train_data_lgbm, test_data_lgbm],
          "drop_rate": 0.1,
          "is_unbalanced": True,
          "max_drop": 50,
          "min_child_samples": 10,
          "min_child_weight": 150,
          "min_split_gain": 0,
          "subsample": 0.9,
          'num_class': 3
          }

clf2 =lgbm.train(params,
                       train_data_lgbm,
                       num_boost_round=15000,
                       valid_sets=[train_data_lgbm, test_data_lgbm],
                       valid_names=['train','val'],
                       feval=lgb_f1_score,
                       verbose_eval=500,
                       early_stopping_rounds=300)
```

```python
joblib.dump(clf2, 'lgbm_final.pkl')
```

```python
lgb_prediction = clf2.predict(X_test)
lgb_prediction = lgb_prediction.argmax(axis = 1)
```

```python
labels = ['low', 'medium', 'high']
plt.figure(figsize=(8,8))
plt.grid(b=False)
cm = confusion_matrix(y_test, lgb_prediction)
plot_confusion_matrix(cm, classes=labels, normalize=True, title='Normalized confusion matrix')
```
